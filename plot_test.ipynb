```python
import matplotlib.pyplot as plt
from scipy.interpolate import splrep, splev
import numpy as np

font = {'name':'Sathu',
        'color':'xkcd:azure',
        'weight':'bold',
        'size':24
       }

x = np.array([0,1,2,3,4,4,5,6,7,8,9,10,11])
y = np.array([0,0,-2,3,6,4,6,7,8,10,8,5,6])
data_fixed = remove_doubles(x,y)

xnew = np.linspace(0,11,100)
ynew = splev(xnew,splrep(data_fixed[0],data_fixed[1]))

plt.figure('fig',figsize=(15,7.5))
plt.plot(xnew,ynew,linewidth=3,c='xkcd:azure')
plt.plot(data_fixed[0],data_fixed[1],'o',c='red')

plt.grid('on')
plt.show()
```

```python
def remove_doubles(x,y):
    xout = [x[0]]
    yout = [y[0]]
    for i in range(1,len(x)):
        if x[i] == x[i-1]:
            yout[-1] = y[i]
        else:
            xout.append(x[i])
            yout.append(y[i])
            
    return [xout,yout]
```
